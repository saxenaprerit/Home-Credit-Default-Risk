```python
def calc_NAs(df, df_name):
    print("Dataframe : ", df_name)
    print("Total NAs", "\n", df.isnull().sum().sort_values(ascending =  False), "\n", "Percentage NAs", "\n", (df.isnull().mean().sort_values(ascending = False)*100))
```
